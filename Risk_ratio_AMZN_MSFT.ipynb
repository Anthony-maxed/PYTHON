```python
import pandas as pd
import numpy as np
import matplotlib as plt
import yfinance as yf
```

```python
#Seleccionamos los tickers
ticker = ["AMZN", "MSFT"]
```

```python
#Descargamos la data en este caso de 2 años
datos = yf.download(ticker, period= "2y") 
datos.head(5)
```

```python
#Solo usamos los precios de cierre y lo ponemos en una nueva variable
precio = datos["Close"]
precio
```

```python
#Hallamos los retornos y borramos la primera fila
returns = precio.pct_change().dropna()
returns
```

```python
#Hallamos el retorno compuesto diario
r_days = returns.shape[0]
returns_day = (returns+1).prod()**(1/r_days) -1
returns_day
```

```python
#Hallamos el retorno compuesto anual
returns_all = (returns_day+1)**252 -1
returns_all 
```

```python
#Ahora hallamos la volatilidad anualizada
volatilidad = returns.std()*(252**0.5)
volatilidad
```

```python
#RATIOS
```

```python
r =returns_all/volatilidad
r
```

```python
riskfree = 0.03
rs =(returns_all-riskfree)/volatilidad
rs
```
